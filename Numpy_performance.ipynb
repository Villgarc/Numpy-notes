## **What should I used Numpy instead of Python lists**
Numpy is symply far more efficient than traditional list. Not only it consumes less memory than lists, but it also provides a mechanism of specifying datatypes. The cells below show a couple of examples the efficiency of numpy arrays if we compare it with Python lists.

```python
from time import time
import matplotlib.pyplot as plt
from typing import Callable
def benchmark(func : Callable):
  times = []
  for i in range(20000):
    start = time()
    func(i)
    end = time()
    if i%500 == 0: times.append(end-start)
  length = [i*500 for i in range(40)]
  plt.plot(length,times)
  plt.xlabel("Number of elements")
  plt.ylabel('Time in seconds')
  plt.title('Benchmark results')
  plt.show()
  
```

```python
def create_list (range_):
  list1 = [i for i in range(range_)] 
  return list1
```

```python
benchmark(create_list)
```

```python
def create_numpy_array(range_): return np.arange(range_)
```

```python
benchmark(create_numpy_array)
```

```python
%%timeit
create_list(100000)
```

```python
%%timeit
create_numpy_array(100000)
```

## **Addition**

```python
def add1(range_):
  list1 = [i for i in range(range_)]
  list1 = [i+5 for i in list1]
```

```python
benchmark(add1)
```

```python
def add2(range_):
  list2 = np.arange(range_)
  list2 += 2 
  return list2
```

```python
benchmark(add2)
```

```python
%%timeit
list1 = [i for i in range(100000)]
list1 = [i+5 for i in list1]
```

```python
%%timeit
list2 = np.arange(100000)
list2 += 2 
```

## **Deletion**

```python
list1 = [i for i in range(100000)]
```

```python
def del1(range_):
  list1 = [i for i in range(range_)]
  del(list1)
```

```python
benchmark(del1)
```

```python
def del2(range_):
  list2 = np.arange(range_)
  del(list2)
```

```python
benchmark(del2)
```

```python
%%timeit
list1 = [i for i in range(100000)]
del(list1)
```

```python
%%timeit
list2 = np.arange(100000)
del(list2)
```

Now, after performing some tests, it is clear that numpy is far more efficient that traditional Python.
